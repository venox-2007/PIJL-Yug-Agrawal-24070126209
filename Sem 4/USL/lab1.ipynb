```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial import distance
```

```python
# Loading Iris dataset
from sklearn.datasets import load_iris
iris=load_iris()
x=iris.data[:5] # Take first 5 rows for demo
print('Sample Data:\n',x)

# Select two sample points
point1=x[0]
point2=x[1]

# Euclidean Distance
print("\nEuclidean:", distance.euclidean(point1, point2))

# Manhattan Distance
print("Manhattan:", distance.cityblock(point1, point2))

# Minkowski Distance (p = 3)
print("Minkowski (p=3):", distance.minkowski(point1, point2, p=3))

# Hamming Distance
print("Hamming:", distance.hamming(point1, point2))

# Cosine Distance
print("Cosine Distance:", distance.cosine(point1, point2))

# Cosine Similarity
print("Cosine Similarity:", 1 - distance.cosine(point1, point2))
```

```python
# Jaccard Distance Demonstration
a = np.array([1, 1, 0, 1, 0])
b = np.array([1, 0, 0, 1, 1])

# Jaccard Distance
jaccard_dist = distance.jaccard(a, b)
print("Jaccard Distance:", jaccard_dist)
```
